```python
from google.colab import files
import pandas as pd

#Upload the CSV file
uploaded = files.upload()

#Command for upload only one CSV file
filename = next(iter(uploaded))

#Read the CSV file into a DataFrame
df = pd.read_csv(filename)

#Display the first few rows of the DataFrame
print(df.head())

```

```python
#Remove unwanted columns from the Data Frame
remove_cols = ['X_COORD_CD', 'Y_COORD_CD', 'Latitude', 'Longitude', 'New Georeferenced Column']

# Check if each column in remove_cols exists in the DataFrame
for col in remove_cols:
    if col not in df.columns:
        print(f"Warning: Column '{col}' not found in DataFrame.")

# Drop only the columns that exist
existing_cols = [col for col in remove_cols if col in df.columns]
df.drop(columns=existing_cols, inplace=True)

print(df.columns)
```

```python
#List all the values in the ARREST_BORO column
df['ARREST_BORO'].unique()
```

```python
#Compare total arrest counts for Kings, Queens and Manhattan boroughs
arrests_borough_kings = df['ARREST_BORO'][df['ARREST_BORO'] == 'K'].count()
arrests_borough_queens = df['ARREST_BORO'][df['ARREST_BORO'] == 'Q'].count()
arrests_borough_manhattan = df['ARREST_BORO'][df['ARREST_BORO'] == 'M'].count()
arrests_borough_bronx = df['ARREST_BORO'][df['ARREST_BORO'] == 'B'].count()
arrests_borough_staten_island = df['ARREST_BORO'][df['ARREST_BORO'] == 'S'].count()

print(arrests_borough_kings)
print(arrests_borough_queens)
print(arrests_borough_manhattan)
print(arrests_borough_bronx)
print(arrests_borough_staten_island)

#Create bar graph of total arrests per borough
arrests_per_borough = [arrests_borough_kings, arrests_borough_queens, arrests_borough_manhattan, arrests_borough_bronx, arrests_borough_staten_island]
boroughs = ['Kings', 'Queens', 'Manhattan', 'Bronx', 'Staten Island']
import matplotlib.pyplot as plt
plt.bar(boroughs, arrests_per_borough)
plt.xlabel('Borough')
plt.ylabel('Total Arrests')
plt.title('Total Arrests per Borough')
plt.show()

```
